```python
# Import pandas
import pandas as pd
```

```python
# Load Data
df = pd.read_excel('Excel//fact.xlsx')
```

```python
# Display the column names
print("\nColumn Names:\n")
print(df.columns)
```

```python
# Check for any missing values in the DataFrame
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
# Display basic information about the DataFrame
print("DataFrame Information:\n")
df.info()
```

```python
# Convert the approval_date column to datetime
df['approval_date'] = pd.to_datetime(df['approval_date'], errors='coerce')

# Verify the conversion
print("DataFrame Information:\n")
df.info()
```

```python
# Drop rows where approval_date is NaT (Not a Time)
df = df.dropna(subset=['approval_date'])

# Verify the number of rows after dropping
print(f"Number of rows after dropping: {len(df)}")
```

```python
print("DataFrame Information:\n")
df.info()
```

```python
# Check for any missing values in the DataFrame
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
# Replace NaN values in qty_used column with 1
df['qty_used'].fillna(1, inplace=True)

# Verify the changes
print(df['qty_used'].isnull().sum())
```

```python
print("DataFrame Information:\n")
df.info()
```

```python
# Convert rto_validation and revenue_month to datetime with only month and year
df['rto_validation'] = pd.to_datetime(df['rto_validation'], errors='coerce').dt.to_period('M')
df['revenue_month'] = pd.to_datetime(df['revenue_month'], errors='coerce').dt.to_period('M')

# Verify the conversion
print(df[['rto_validation', 'revenue_month']].head())
```

```python
# Drop the specified columns
df = df.drop(columns=['po', 'invoice', 'rev_x_val','profit'])

# Verify the change
print(df.columns)
```

```python
# Rename the columns
df.rename(columns={
    'jobcode': 'Job_Code',
    'request_date': 'Request_Date',
    'ihs_id': 'Site_ID',
    'job_type': 'Job_Type',
    'fault': 'Job_Description',
    'qty': 'Qty_Approved',
    'unit': 'Approved_Unit_Price',
    'total_approved': 'Total_Approved',
    'approval': 'Customer_Name',
    'approval_date': 'Approval_Date',
    'job_status': 'Job_Status',
    'closure_date': 'Closure_Date',
    'payment_ref': 'Execution_Payment_Reference',
    'executor': 'Execution_Type',
    'qty_used': 'Qty_Used',
    'unit_used': 'Unit_Price',
    'total_exepense': 'Total_Expense',
    'rto_validation': 'Customer_Validation_Month',
    'revenue_month': 'Revenue_Month'
}, inplace=True)

# Verify the changes
print(df.columns)
```

```python
import matplotlib.pyplot as plt

job_type_counts = df['Job_Type'].value_counts()
job_type_counts.plot(kind='bar', title='Job Type Distribution')
plt.xlabel('Job Type')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
common_job_types = df['Job_Description'].value_counts().head(10)
common_job_types.plot(kind='bar', title='Top 10 Common Job Descriptions')
plt.xlabel('Job Description')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
df['Request_Date'] = pd.to_datetime(df['Request_Date'])
monthly_trend = df['Request_Date'].dt.to_period('M').value_counts().sort_index()
monthly_trend.plot(kind='line', title='Request Trend by Months')
plt.xlabel('Month')
plt.ylabel('Number of Requests')
plt.show()
```

```python
import seaborn as sns

# Filter data for the year 2023
df_2023 = df[df['Request_Date'].dt.year == 2023]

# Extract month from the request date
df_2023['Request_Month'] = df_2023['Request_Date'].dt.month

# Group by month and count the number of requests
monthly_job_counts_2023 = df_2023.groupby('Request_Month').size().reset_index(name='Job_Count')

# Create a correlation plot
sns.lmplot(x='Request_Month', y='Job_Count', data=monthly_job_counts_2023)
plt.title('Correlation between Month and Job Count (2023)')
plt.xlabel('Month')
plt.ylabel('Job Count')
plt.show()
```

```python
# Extract day from the request date
df['Request_Day'] = df['Request_Date'].dt.day

# Group by day and count the number of requests
daily_job_counts = df.groupby('Request_Day').size().reset_index(name='Job_Count')

# Create a correlation plot
sns.lmplot(x='Request_Day', y='Job_Count', data=daily_job_counts)
plt.title('Correlation between Days and Job Count')
plt.xlabel('Day of the Month')
plt.ylabel('Job Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a crosstab of Job_Type and Request_Day
crosstab = pd.crosstab(df['Request_Day'], df['Job_Type'])

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Heatmap of Job Type by Day of the Month')
plt.xlabel('Job Type')
plt.ylabel('Day of the Month')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for the year 2023
df_2023 = df[df['Request_Date'].dt.year == 2023]

# Extract month from the request date
df_2023['Request_Month'] = df_2023['Request_Date'].dt.month

# Create a crosstab of Job_Type and Request_Month
crosstab = pd.crosstab(df_2023['Request_Month'], df_2023['Job_Type'])

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Heatmap of Job Type by Month in 2023')
plt.xlabel('Job Type')
plt.ylabel('Month')
plt.show()
```
